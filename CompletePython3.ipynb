# Git พื้นฐาน

Git's files status
1. Untrack File    คือ ไฟล์ที่พึ่งเปลี่ยน แต่ Git ยังไม่รู้ว่าเปลี่ยน
2. Unstage File    คือ ไฟล์ที่พึ่งเปลี่ยนแปลง แต่ยังไม่ถูก add เพื่อ commit
3. Stage File      คือ ไฟล์ที่ถูก add แล้ว พร้อมจะ commit
4. Deleted File    คือ ไฟล์ที่ถูกลบไปแล้ว


Commit คือ ส่ง Stage File ไปเก็บไว้ใน Git

## คำสั่ง Git

1. ดาวน์โหลด Git

## Special Symbols Used for passing arguments

1. ***args** (Non-Keyword Arguments) รับค่าแบบ Tuple

เช่น


````python
def myFun(*argv):
    for arg in argv:
        print(arg)
myFun(*('I\'m', '12', 'yo'))
````

````python
def myFun(*args):
    for argg in args:
        print(argg)
arg_tuple = ('I\'m', '12', 'yo')
myFun(*arg_tuple)
````

````python
def myFun(*args):
    print(args[0])
    print(args[1])
    print(args[2])
arg_tuple = ('I\'m', '12', 'yo')
myFun(*arg_tuple)
````

2. ****kwargs** (Keyword Arguments) รับค่าแบบ Dictionary
เช่น




````python
def myFun(**kwargs):
    for value in kwargs.values():
      print(value)
myFun(**{"1": "I'm", "age": "12", "year": "yo"})
````

````python
def myFun(**kwargs):
    for value in kwargs.values():
      print(value)
a = {"1": "I'm", "age": "12", "year": "yo"}
myFun(**a)
````

````python
class car():

    def __init__(self, **kwargs):
        self.speed = kwargs['s']
        self.color = kwargs['c']

audi = car(s=200, c='red')
print(audi.color)
````

* ***Catch***: ดักเอาไว้ หาทางแก้ให้โปรแกรม

````python
import sys
randomList = ['a', 0, 2]
for entry in randomList:
  try:
    print("The entry is", entry)
    r = 1/int(entry)
    break
  except:
    print("Oops!", sys.exc_info()[0], "occured.")
    print("Next ecntry.")
    print()
print("The reciprocal of", entry, "is", r)
````

````python
import csv
#ส่วนการเปิดไฟล์ CSV ที่ชื่อ employee_file.txt
with open('employee_file.csv', mode='w') as csv_file:
    #ส่วนการกำหนดการเขียนไฟล์
    fieldnames = ['Name', 'Department', 'Birthday month']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    #ส่วนการเพิ่มข้อมูล
    writer.writeheader()
    writer.writerow({'Name': 'John Smith', 'Department': 'Accounting', 'Birthday month': 'November'})
    writer.writerow({'Name': 'Erica Meyers', 'Department': 'IT', 'Birthday month': 'March'})
````

### โปรแกรมคำนวณปลา รถพุ่มพวงของลุง

````python
from tkinter import * #All ใน main
from tkinter import ttk, messagebox #อยู่ไฟล์แยก

GUI = Tk()
GUI.title('โปรแกรมคำนวณราคาปลา รถพุ่มพวง')
GUI.geometry('700x600')

bg = PhotoImage(file='car.png')

BG = Label(GUI, image=bg)
BG.pack()

L = Label(GUI, text='กรอกน้ำหนักปลา(กก.)', font=(None,20))
L.pack()

v_quantity = String()
Entry = ttk.Entry(GUI, textvariable=v_quantity, font(None,20))
Entry.pack()

def Cal():
    try:
        quan = float(v_quantity.get())
        calc = quan*39 #39 บาท/กก.
        messagebox.showinfo('ราคาทั้งหมด', 'ราคาปลาทั้งหมด {} บาท'.format(calc))
        v_quantity.set('')
        Entry.focus()
    except:
        messagebox.showwarning('กรอกผิด', 'กรุณากรอกตัวเลขเท่านั้น')
        v_quantity.set('1')
        Entry.focus()

B = ttk.Button(GUI, text='คำนวณ', command=Cal)
B.pack(ipadx=30, ipady=20, pady=20)

GUI.mainloop()
````

# Object-Oriented Programming

**มองเป็นวัตถุที่สามารถผลิตออกมาซ้ำได้เรื่อย ๆ แค่ปรับแม่พิมพ์ใหม่ให้ตรงตามต้องการ ไม่ต้องเริ่มต้นสร้างขึ้นมาใหม่ทั้งหมด**

## 3 ส่วนสำคัญ

1. Identity: ระบุตัวตน ว่าเป็นตัวไหน
2. State: สถานะ, ลักษณะ, หรือคุณสมบัติของสิ่งของต่าง ๆ เช่น ผมยาว สูงมาก
3. Behavior: พฤติกรรม เช่น เดินได้ วิ่งได้

## 4 คำสำคัญใน OOP

1. Abstract: นามธรรม จับต้องไม่ได้ แค่กำหนดคร่าว ๆ แต่ยังไม่จำเป็นต้องสร้างขึ้นมาจริง ๆ
2. Encapsulation: ห่อหุ้มปกปิกข้อมูลเอาไว้ไม่ให้เข้าถึงได้จากภายนอกโดยตรง
3. Inheritance: การสืบทอด State จากแม่แบบ
4. Polymorphism: การมีได้หลากหลายรูปแบบ ขึ้นอยู่กับการใช้งาน

## Class, Attribure, Behavior, & Object

* Class: สร้างพิมพ์เขียวสำหรับสร้างวัตถุ เช่น พิมพ์เขียวบ้านในหมู่บ้านจัดสรร
* Attribute: ตัวแปรที่เก็บข้อมูลสถานะ, ลักษณะ, หรือคุณสมบัติ เช่น ขนาดพื้นที่ของบ้าน
* Behavior: พฤติกรรมที่ Class สามารถทำงานได้ เช่น บ้านสามารถเปิดประตูได้
* Object: สิ่งที่ถูกสร้างขึ้นมาจากพิมพ์เขียวของ Class เช่น บ้านแต่ละหลังในหมูบ้านจัดสรร

Inherite แบบกำหนด Attribute ที่ Class ลูก

````python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class SpecialStudent(Student):
    def __init__(self, name, age, parent):
        super().__init__(name, age)
        self.parent = parent

print(SpecialStudent('a', 10, 'b'))
````

````python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "https://www.sfcinemacity.com/"

webopen = urlopen(url)
html_page = webopen.read().decode('utf-8')
webopen.close()

data = BeautifulSoup(html_page, 'html.parser')

title = data.find_all('h1',{'class':'heading'})
search = data.find_all('li',{'class':'active'})

print(search[0].text)
print(title[0].text)
````
